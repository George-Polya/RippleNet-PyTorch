```python
with open("./train.txt", "r") as f:
    train = f.readlines()
train = [line.replace("\n","") for line in train]
train = [line.split() for line in train]

with open("./test.txt", "r") as f:
    test = f.readlines()
test = [line.replace("\n","") for line in test]
test = [line.split() for line in test]

train[:5]


```

```python
user_item_dict = dict()
for line in train:
    key = line[0]
    user_item_dict[key] = list()
    
```

```python
for line in train:
    key = line[0]
    for element in line[1:]:
        user_item_dict[key].append(element)
```

```python
user_item_dict
```

```python
for line in test:
    key = line[0]
    for element in line[1:]:
        user_item_dict[key].append(element)
```

```python
with open("./ratings.dat", "w") as f:
    for user_idx, item_list in user_item_dict.items():
        for item_idx in item_list:
            line = [user_idx, item_idx, '5']
            # print("::".join(line))
            f.write("::".join(line)+"\n")
    
```
